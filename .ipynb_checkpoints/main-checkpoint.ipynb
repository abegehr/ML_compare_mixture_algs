```python
import itertools
import numpy as np
from scipy import linalg
from sklearn import datasets, mixture
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
# data generator settings

n_samples = 1500
n_features = 2
centers = 4
cluster_std = [np.random.ranf()*2 for i in range(centers)]
center_box = (-10, 10)
shuffle = True
random_state = np.random.randint(0, 999)

print("cluster_std: ", cluster_std)
print("random_state: ", random_state)
```

```python
# generate data

data = datasets.make_blobs(n_samples=n_samples, n_features=n_features, centers=centers, cluster_std=cluster_std, center_box=center_box, shuffle=shuffle, random_state=random_state)

print("data: ", data)
```

```python
# plot data

X = data[0]
c = data[1]

xs = X[:, 0]
ys = X[:, 1]


fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111)
ax.scatter(xs, ys, 20, c=c)
```

```python
def color(i):
    colors = ['navy', 'c', 'cornflowerblue', 'gold', 'darkorange', 'navy', 'c', 'cornflowerblue', 'gold', 'darkorange', 'navy', 'c', 'cornflowerblue', 'gold', 'darkorange']
    return colors[i]
```

```python
# update_result function
# updates data, gaussian mixtures, and classification

def update_results(scat, ells, mix):
    print("called!")
    Y_ = mix.predict(X)
    means = mix.means_
    covariances = mix.covariances_
    
    # update points color
    scat.set_array(Y_)
    
    # update ellipses
    for i, (mean, covar) in enumerate(zip(
            means, covariances)):
        v, w = linalg.eigh(covar)
        v = 2. * np.sqrt(2.) * np.sqrt(v)
        u = w[0] / linalg.norm(w[0])
        # as the DP will not use every component it has access to
        # unless it needs it, we shouldn't plot the redundant
        # components.
        if not np.any(Y_ == i):
            continue

        # Plot an ellipse to show the Gaussian component
        angle = np.arctan(u[1] / u[0])
        angle = 180. * angle / np.pi  # convert to degrees
        ells[i].xy = mean
        ells[i].width = v[0]
        ells[i].height = v[1]
        ells[i].angle = 180. + angle
    
    return ells + [scat]
        
```

```python
# animate results function

def animate_bgm(parameters, X, start, end, step):
    
    fig = plt.figure(figsize=(12, 12))
    ax = fig.add_subplot(111)
    
    scat = ax.scatter(X[:, 0], X[:, 1], 20)
    ells = []
    for i in range(parameters['n_components']):
        ell = mpl.patches.Ellipse([0,1], 1, 1, 1, color=color(i), animated=True)
        ell.set_clip_box(ax.bbox)
        ell.set_alpha(0.5)
        
        ax.add_patch(ell)
        ells.append(ell)
    
    def init_func():
        return ells + [scat]

    def update(frame, parameters, X, scat, ells):
        print("frame: ", frame)
        bgm = mixture.BayesianGaussianMixture(**parameters, max_iter=frame).fit(X)
        return update_results(scat, ells, bgm)
    
    return FuncAnimation(fig, update, fargs=(parameters, X, scat, ells), frames=range(end, step), init_func=init_func, blit=True)
    
```

```python
# BayesianGaussianMixture
# settings
parameters = {
    'n_components': 10,
    'random_state': np.random.randint(0, 999)
}

print("random_state: ", parameters['random_state'])


# run

anim = animate_bgm(parameters, X, 1, 100, 10)

HTML(anim.to_jshtml())
```
